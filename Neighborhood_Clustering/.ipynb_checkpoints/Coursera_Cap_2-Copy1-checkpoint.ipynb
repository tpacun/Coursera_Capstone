```python
#select most recent real estate price and neighborhood names
neighborhoods = boston[['RegionName', 'City', 'State', '2020-05-31']].reset_index(drop= True)
neighborhoods.rename(columns = {'RegionName': 'Neighborhoods', '2020-05-31': 'RE_price'}, inplace = True)
neighborhoods
```

```python
#google api data
google_key = 'AIzaSyAJmiz1eaIhZK1J4BH0NsN4640dyHth38Q'
```

```python
#function to get coordinates
def get_coord(apikey, address):
    try:
        url = 'https://maps.googleapis.com/maps/api/geocode/json?key={}&address={}'.format(apikey, address)
        response = requests.get(url).json()
        geo_location = response['results'][0]['geometry']['location']
        lat, lng = geo_location['lat'], geo_location['lng']
        return [lat, lng]
    except:
        return [None, None]
```

```python
# get lat, long of boston neighborhoods
neighborhood_coordinates = []
for hood, city, state in zip(neighborhoods['Neighborhoods'], neighborhoods['City'], neighborhoods['State']):
    coordinates = get_coord(google_key, hood + city + state)
    neighborhood_coordinates.append(coordinates)

print(len(neighborhood_coordinates))
```

```python
#parse out data to indivual lists
hood_lat, hood_lng = [], []

for i in range(len(neighborhood_coordinates)):
    hood_lat.append(neighborhood_coordinates[i][0])
    hood_lng.append(neighborhood_coordinates[i][1])

print(hood_lat, hood_lng)
```

```python
#add into df
neighborhoods['Lat'] = hood_lat
neighborhoods['Long'] = hood_lng
neighborhoods.head()
```

```python
#map boston
map_boston = folium.Map(location = boston_center, zoom_start = 12)
folium.Marker(boston_center, popup = 'Boston Center').add_to(map_boston)
map_boston
```

```python
#feature group of neighborhoods
fg_neighborhoods = folium.map.FeatureGroup()

#add neighborhoods to feature group
for lat, lon, neigh in zip(neighborhoods['Lat'], neighborhoods['Long'], neighborhoods['Neighborhoods']):
    fg_neighborhoods.add_child(
        folium.CircleMarker(
            [lat, lon],
            radius = 8,
            color = 'blue',
            fill = True,
            popup = neigh))
    
#add group and display map
map_boston.add_child(fg_neighborhoods)
```
